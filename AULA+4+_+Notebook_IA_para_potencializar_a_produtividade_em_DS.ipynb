## Tend√™ncia de produ√ß√£o por fonte

ü§ñ **Prompt:**

Quero que voc√™ atue como um cientista de dados e codifique para mim. Estou desenvolvendo um projeto de Data Science focado no tema de consumo e produ√ß√£o de energia el√©trica pelo mundo. Agora preciso verificar a tend√™ncia de produ√ß√£o de energia por 3 tipos diferentes de fontes de energia.

No dataframe "resumo_mundial" existem 3 colunas que se caracterizam por serem s√©ries temporais chamadas "producao_total_de_combustiveis_fosseis", "producao_de_energia_nuclear" e "producao_total_de_energia_renovavel".

Existe outra coluna em "resumo_mundial" chamada "data" que tem datas em intervalos fixos em cada linha, mostrando o final de todos os meses durante os anos de 1973 e 2022.

N√£o altere o dataframe "resumo_mundial" para ajustar os dados, se for preciso, crie um novo dataframe.

Crie um c√≥digo que plote a tend√™ncia dos valores de cada tipo de produ√ß√£o ao longo dos anos. Utilize a biblioteca "statsmodels" para encontrar a tend√™ncia das s√©ries com "seasonal_decompose".

Ao fim, crie uma plotagem com a s√©rie temporal de cada um dos tipos de produ√ß√£o com plotly. Adicione t√≠tulo e legenda que fa√ßam sentido.

````python
import statsmodels.api as sm

# 2. Criar um novo DataFrame com as colunas de interesse
producao_tipos = resumo_mundial[["data", "producao_total_de_combustiveis_fosseis", "producao_de_energia_nuclear", "producao_total_de_energia_renovavel"]]

# 3. Utilizar seasonal_decompose para decompor as s√©ries temporais
producao_tipos.set_index("data", inplace=True)
decomposed_combustiveis = sm.tsa.seasonal_decompose(producao_tipos["producao_total_de_combustiveis_fosseis"], model='additive')
decomposed_nuclear = sm.tsa.seasonal_decompose(producao_tipos["producao_de_energia_nuclear"], model='additive')
decomposed_renovavel = sm.tsa.seasonal_decompose(producao_tipos["producao_total_de_energia_renovavel"], model='additive')

# 4. Criar a plotagem usando plotly
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=producao_tipos.index,
    y=decomposed_combustiveis.trend,
    mode='lines',
    name='Produ√ß√£o de Combust√≠veis F√≥sseis (Tend√™ncia)'
))


fig.add_trace(go.Scatter(
    x=producao_tipos.index,
    y=decomposed_nuclear.trend,
    mode='lines',
    name='Produ√ß√£o de Energia Nuclear (Tend√™ncia)'
))


fig.add_trace(go.Scatter(
    x=producao_tipos.index,
    y=decomposed_renovavel.trend,
    mode='lines',
    name='Produ√ß√£o de Energia Renov√°vel (Tend√™ncia)'
))

fig.update_layout(
    title='Tend√™ncia de Produ√ß√£o de Energia por Tipo',
    xaxis_title='Data',
    yaxis_title='Produ√ß√£o de Energia',
    legend_title='Tipo de Produ√ß√£o',
    plot_bgcolor='white'  # Definir o fundo do gr√°fico como branco
)

fig.show()
````
